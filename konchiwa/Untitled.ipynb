```python
import requests
from bs4 import BeautifulSoup

url = "https://www.journals.uchicago.edu/toc/hr/current" 
HP = requests.get(url)
soup = BeautifulSoup(HP.text, "html.parser")

#論文のタイトル
papers = soup.find_all(class_="hlFld-Title")
for paper in papers:
    titles = paper
    print(titles)
```
